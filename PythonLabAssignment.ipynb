```python
# Question 1: To implement all arithmetic,relational,logical operators in python

x = int(input("Enter First Number:"))
y = int(input("Enter Second Number:"))

def ArithmaticOp(x , y):
    print("Addition Operation",x,"+",y,":",(x+y))
    print("Subtraction Operation",x,"-",y,":",(x-y))
    print("Multiplication Operation",x,"*",y,":",(x*y))
    print("Division Operation",x,"/",y,":",(x/y))
    print("Modulus Operation",x,"%",y,":",(x%y))
    print("Floor division Operation",x,"//",y,":",(x//y))
    print("Exponent Operation",x,"**",y,":",(x**y))
    
def RelationalOp(x , y):
    print("Greater than Operation",x,">",y,":",(x>y))
    print("Less than Operation",x,"<",y,":",(x<y))
    print("Equal to Operation",x,"==",y,":",(x==y))
    print("Not equal to Operation",x,"!=",y,":",(x!=y))
    print("Greater Than or Equal to Operation",x,">=",y,":",(x>=y))
    print("Less Than or Equal to Operation",x,"<=",y,":",(x<=y))
    
def LogicalOp(x , y):
    print("AND Operation",x,"> 10 and",y,"<100 :",(x>10) and (y<100))
    print("OR Operation",x,"< 200 or",y,"> 50:",(x<200) or (y>50))
    print("NOT Operation","not",x,"<",y,":",not(x<y))

print("Options=> \n1)To implement arithmetic operators\n2)To implement relational operators\n3)To implement logical operators")
print("4)To implement all operators")
op= int(input("Enter Choice:"))

if op == 1:
  ArithmaticOp(x,y)

elif op == 2:
  RelationalOp(x,y)

elif op == 3:
  LogicalOp(x,y)

elif op == 4:
  ArithmaticOp(x,y)
  RelationalOp(x,y)
  LogicalOp(x,y)

else:
    print("Wrong Option Entered,Please Try again!!!")
```

```python
#Question 3:To check 1) Greatest of two Numbers
#                    2) Greatest of three Numbers

def GreatCheck (a,b,c = 0):
    if c > b:
        b = c

    if a > b:
        return a
    else:
        return b
    
print("Options=> \n1)Greatest of Two Numbers \n2)Greatest of Three Numbers")
op= int(input("Enter Choice:"))

if op == 1:
    num1 = int(input("Enter 1st Number:"))
    num2 = int(input("Enter 2nd Number:"))

    print("Resultant Greatest of 2 numbers:",GreatCheck(num1,num2))

if op == 2:
    num1 = int(input("Enter 1st Number:"))
    num2 = int(input("Enter 2nd Number:"))
    num3 = int(input("Enter 3rd Number:"))

    print("Resultant Greatest of 3 numbers:",GreatCheck(num1,num2,num3))

else:
    print("Wrong Option Entered,Please Try again!!!")
```

```python
#Question 8:To find number is Palindrome or not

num = int(input("Enter a Number:"))
temp = num
newNum =0

while temp > 0:

    newNum *=10
    newNum += temp % 10
    temp //=10


if newNum == num:
    print("Entered Number",num,"is a Palindrome")

else:
    print("Entered Number",num, "is not a Palindrome")
```

```python
#Question 9:To check number is armstrong number or not

num = int(input("Enter a Number:"))
temp = num
newNum =0

while temp > 0:

    newNum += (temp % 10)**len(str(num))
    temp //=10


if newNum == num:
    print("Entered Number",num,"is a Armstrong")

else:
    print("Entered Number",num, "is not a Armstrong")
```

```python
#Question 19:To fibonacci series using recursion

def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))

nterms = int(input("Enter a Number: "))

if nterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence:")
   for i in range(nterms):
       print(recur_fibo(i))
```

```python
#Question 21:To find sum of first n natural numbers using recursion

def recurSum(n):
    if n <= 1:
        return n
    return n + recurSum(n - 1)

num = int(input("Enter a Number: "))

print("Sum of first",num,"Natural Numbers:",recurSum(num))
```

```python
#Question 25:Write python program to add two matrices

matrix1 = []
matrix2 = []
result = []
s = int(input("Enter the Matrix size:"))
print("Enter Values for Matrix 1:")

for i in range(s):          
    a =[]
    res =[]
    for j in range(s):      
         a.append(int(input()))
         res.append(0)
    matrix1.append(a)
    result.append(res)

print("Enter Values for Matrix 2:")
for i in range(s):          
    a =[]
    for j in range(s):      
        a.append(int(input()))
    matrix2.append(a)

for i in range(s):
   for j in range(s):
       result[i][j] = matrix1[i][j] + matrix2[i][j]

print("Matrix 1:")
for m1 in matrix1:
     print(m1)

print("Matrix 2:")
for m2 in matrix2:
     print(m2)

print("Result:")
for r in result:
   print(r)
```

```python
#Question 26:Write python program to Multiply two matrices

matrix1 = []
matrix2 = []
result = []
r1 = int(input("Enter the row size for Matrix 1:"))
c1 = int(input("Enter the column for Matrix 1:"))
print("Enter Values for Matrix 1:")

for i in range(r1):          
    a =[]
    for j in range(c1):      
        a.append(int(input()))
    matrix1.append(a)

r2 = int(input("Enter the row size for Matrix 2:"))
c2 = int(input("Enter the column for Matrix 2:"))
print("Enter Values for Matrix 2:")
for i in range(r2):          
    a =[]
    for j in range(c2):      
        a.append(int(input()))
    matrix2.append(a)

for i in range(r1):
    res =[]
    for j in range(c2):
        res.append(0)
    result.append(res)

for i in range(r1):
   for j in range(c2):
       for k in range(r2):
           result[i][j] += matrix1[i][k] * matrix2[k][j]

print("Matrix 1:")
for m1 in matrix1:
     print(m1)

print("Matrix 2:")
for m2 in matrix2:
     print(m2)

print("Result:")
for r in result:
   print(r)
```

```python
#Question 27:Write python program to Transpose a matrix

matrix = []
r = int(input("Enter the row size:"))
c = int(input("Enter the column size:"))
print("Enter Values for Matrix:")

for i in range(r):          
    a =[]
    for j in range(c):      
      a.append(int(input()))
    matrix.append(a)

result = [[matrix[j][i] for j in range(r)] for i in range(c)]

print("Matrix:")
for m in matrix:
     print(m)

print("Result:")
for r in result:
     print(r)
```

```python
#Question 32:Write a python program to identify duplicates in a list

n = int(input("Enter No of inputs:"))

list = []
dup = []

for i in range(n):
    list.append(int(input()))

for i in range(n):
    for j in range(i+1,n):
        if list[i] == list[j] and list[i] not in dup:
            print("Number",list[i],"is Duplicate in list")
            dup.append(list[i])
            break
```

```python
#Question 42:Write a python program to count occurences of each word in a given sentence

s = input("Enter String:")
ref = {}

for text in s:
    text = text.strip()
    text = text.lower()
    words = text.split(" ")

    for word in words:

        if word in ref:
            ref[word] = ref[word] + 1

        else:
            ref[word] = 1

for key in ref.keys():
    print(key, ":", ref[key])
```
